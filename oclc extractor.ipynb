```python
import pymarc
from pymarc import Record, Field
from pymarc import MARCReader
from tqdm import tqdm
import csv
import re
import pandas as pd
```

```python
def extract_oclc(MARCfile, headerrow):
    oclc_loc = input("Is the OCLC# in the 001 or 035? ")
        
    with open(MARCfile, 'rb') as f:
        
        reader = MARCReader(f, to_unicode=True, force_utf8=True)
        results = []
        
        for record in tqdm(reader, desc="Progress", unit=" records"):
        
            oclc = ""
            if oclc_loc == '035':
                if record['035']:
                    f035 = record.get_fields('035')[0]
                    if f035.get_subfields('a'):
                        subfa = f035.get_subfields('a')[0]
                        if 'OCoLC' in subfa:
                            oclc = subfa
                            row = [oclc]
                    
                    df2 = pd.DataFrame([row], columns=headerrow)
                    results.append(df2)
            else:
                if record['001']:
                    oclc = record.get_fields('001')[0]
                    oclc_pref = re.sub("=001  ", "(OCoLC)", str(oclc))     
                    row = [oclc_pref]
                    
                df2 = pd.DataFrame([row], columns=headerrow)
                results.append(df2)
        
        return pd.concat(results)
    
def csv_rename(MARCfile):
    fcsv = str(MARCfile)
    fcsv = fcsv.replace(".mrc", "_oclcs.csv")
    return fcsv
                                       
def pkl_rename(MARCfile):
    fpkl = str(MARCfile)
    fpkl = fpkl.replace(".mrc", "_oclcs.pkl")
    return fpkl
    
def main():
    MARCfile = input("Enter the MARC filename: ")
    headerrow = ["OCLC"]
    oclc_df = extract_oclc(MARCfile, headerrow)
    fpkl = pkl_rename(MARCfile)
    oclc_df.to_pickle(fpkl)
    print('Pickled! Converting to csv...')
    csvname = csv_rename(MARCfile)
    oclc_df.to_csv(csvname, encoding='utf8', index=False)
    print('Pkl to csv complete')

main()
```
